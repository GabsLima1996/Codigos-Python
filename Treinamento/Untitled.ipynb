```python
# Importe de bibliotecas

# Pandas é uma biblioteca para trabalhar com textos 
import pandas as pd
# numpy é uma biblioteca para trabalhar com numeros
import numpy as np
# seaborn é uma biblioteca para trabalhar com graficos
import seaborn as sns
from IPython.display import display
```

```python
# sep é um tipo de separador EX: (; . ..)
df = pd.read_csv('winequality-red.csv', sep=';',quotechar='"')
```

```python
df
```

```python
# describe mostra a descrião do conjunto de dados
display(df.describe(include='all'))
```

```python
# Comando info é para mostrar dados referentes aos dados 
df.info()
```

```python
# Comando .sum é para saber quantos campos nulos existem 
df.isnull().sum()
```

```python
# Comando head mostra as cinco primeiras linha da base
df.head()
```

```python
# Comando groupby agrupa por colunas / Comando count faz a conta da variavel escolhida EX: 
df1 = df.groupby(['quality']).count()
df1
```

```python
# Comando groupby agrupa por colunas / Comando .sum faz a soma da variavel escolhida EX:
df2 = df.groupby(['quality']).sum()
df2 
```

```python
# Comando merge é para unir duas tabelas distintas 
Count = pd.merge(df1[['fixed acidity']], df2[['fixed acidity']], on='quality', how='left')
# Comando fillna é para eliminar casas null 
Count = Count.fillna(0)
# Comando round é para arredondar as casas decimais 
# Count['perc_quality'] =  # Esse comando cria uma nova coluna na base 
Count['perc_quality'] = round((Count['fixed acidity_x'] / Count['fixed acidity_y']) * 100,0)
Count['FODA-SE'] = 'FODA-SE'
Count
```

```python
# Comando Count.plot é para gerar graficos com escala
Count.plot(x ='fixed acidity_x', y ='fixed acidity_y',kind='scatter',color='r')
```

```python
# Comando %matplotlib para gerar graficos 
%matplotlib notebook
Count['fixed acidity_y'].plot()
```

```python
# Comando .hist é para gerer um histograma com a coluna selecionada 
# Comando bins é para agrupar pela quantidade desejada 
%matplotlib notebook
Count['fixed acidity_y'].hist(bins =1000, color ='red')
```
